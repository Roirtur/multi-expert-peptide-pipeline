# Démo du Chemist Agent

Ce notebook montre comment utiliser le `ChemistAgent` pour valider et filtrer des peptides.
Nous devons d'abord configurer le chemin d'accès pour que Python puisse trouver nos modules.

```python
import sys
import os
from peptide_pipeline.chemist import ChemistAgent

project_root = os.path.abspath(os.path.join(os.getcwd(), "../.."))

if project_root not in sys.path:
    sys.path.append(project_root)

print(f"Project root added to path: {project_root}")

# Initialisation de l'agent
agent = ChemistAgent()
print("ChemistAgent initialized successfully!")
```

```python
# Test de validité
peptides_test = ["ACDEF", "INVALIDX", "", "GHIKL"]
validity = agent.check_validity(peptides_test)

for pep, is_valid in zip(peptides_test, validity):
    print(f"Peptide '{pep}': {'Valid' if is_valid else 'Invalid'}")
```

```python
# Calcul des propriétés
props = agent.calculate_properties(["ACDEF", "GHIKL"])
for p, prop in zip(["ACDEF", "GHIKL"], props):
    print(f"\nProperties for {p}:")
    for k, v in prop.items():
        print(f"  {k}: {v}")
```

```python
# Filtrage
constraints = {
    "size": {"min": 3, "max": 10},
    "net_charge": {"min": -2.0, "max": 2.0}
}
filtered = agent.filter_peptides(["ACDE", "GGHK", "AHHGGGGAGGHHH"], constraints)
print(f"\nFiltered peptides: {filtered}")
```
