# Démo du Chemist Agent

```python
import sys
import os
root = os.path.abspath("../")
if root not in sys.path:
    sys.path.append(root)
```

```python
from peptide_pipeline.chemist.chemist_agent import ChemistAgent
from peptide_pipeline.chemist.chemist_agent.config_chemist import ChemistConfig, RangeTarget

# Initialisation de l'agent
chemist_config = ChemistConfig(
    ph=7.0,
    length=RangeTarget(min=8.0, max=14.0, target=10.0),
    molecular_weight=RangeTarget(min=1200.0, max=2000.0, target=float(1600.0)),
    logp=RangeTarget(min=-3.0, max=3.0, target=float(0.0)),
    net_charge=RangeTarget(min=2.0, max=8.0, target=float(5.0)),
    isoelectric_point=RangeTarget(min=9.0, max=14.0, target=float(12.0)),
    hydrophobicity=RangeTarget(min=-2.0, max=3.0, target=float(0.0)),
)
agent = ChemistAgent(chemist_config)
print("ChemistAgent initialized successfully!")
```

```python
# Test de validité
peptides_test = ["ACDEF", "INVALIDX", "XXAAH", "GHIKL", "A", "COH", 'CFFFFF']
agent.evaluate_peptides(peptides_test)
```

```python
# Filtrage
top_peptides = agent.get_top_filtered_peptides(peptides_test, topK=3)
print("Top peptides:", top_peptides)
```
